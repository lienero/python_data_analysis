```python
# 피처 각각에 대한 히스토그램을 출력하는 함수를 만듭니다..
def plot_hist_each_column(df):
    # rcParams을 이용하여 차트 그림(figure)의 기본 설정을 지정할 수 있습니다. 
    # "figure.figsize" 그림(figure)의 크기. (가로,세로) 인치 단위
    plt.rcParams['figure.figsize'] = [20, 16]
    fig = plt.figure(1)
    # df의 열 개수 만큼의 subplot을 출력합니다.
    for i in range(len(df.columns)):
        ax = fig.add_subplot(5, 5, i+1)
        plt.hist(df[df.columns[i]], bins=50)
        ax.set_title(df.columns[i])
    plt.show()    
```

```python
# 피처 각각에 대한 히스토그램을 출력합니다.
plot_hist_each_column(picher_features_df)
```

```python
# 판다스 형태로 정의된 데이터를 출력할 때 scientific-notation이 아닌 float 모양으로 출력되게 해줍니다.
pd.options.mode.chained_assignment = None
```

```python
# 피처 각각에 대한 스케일링을 수행하는 함수를 정의합니다.
def standard_scaling(df, scale_columns):
    for col in scale_columns:
        series_mean = df[col].mean()
        series_std = df[col].std()
        df[col] = df[col].apply(lambda x: (x-series_mean)/series_std)
    return df
```

```python
# 피처 각각에 대한 scaling을 수행합니다.
scale_columns = ['승', '패', '세', '홀드', '블론', '경기', '선발', '이닝', '삼진/9',
       '볼넷/9', '홈런/9', 'BABIP', 'LOB%', 'ERA', 'RA9-WAR', 'FIP', 'kFIP', 'WAR', '연봉(2017)']
picher_df = standard_scaling(picher, scale_columns)
```

```python
picher_df = picher_df.rename(columns={'연봉(2018)': 'y'})
picher_df.head(5)
```

```python
# 팀명 피처를 원-핫 인코딩으로 변환합니다.
# pd.get_dummies 처리 : 결측값을 제외하고 0과 1로 구성된 더미값이 만들어진다.
team_encoding = pd.get_dummies(picher_df['팀명'])
picher_df = picher_df.drop('팀명', axis=1)
picher_df = picher_df.join(team_encoding)
```

```python
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
# 학습 데이터와 테스트 데이터로 분리합니다.
# 제외선택: .columns.difference([])를 이용하여 제외할 컬럼 나열
X = picher_df[picher_df.columns.difference(['선수명','y'])]
y = picher_df['y']
# train_test_split 함수는 전체 데이터셋 배열을 받아서 랜덤하게 test/train 데이터 셋으로 분리해주는 함수이다.
# Dataset을 나눌 때 test_size 옵션으로 Train, Test의 비율을 설정할 수 있고, random_state로 seed 값을 지정할 수 있다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
# 회귀 분석 계수를 학습합니다. (회귀 모델 학습)
# linear_model.LinearRegression() : LinearRegression 모델을 생성
lr = linear_model.LinearRegression()

# fit() 메서드는 선형 회귀 모델에 필요한 두 가지 변수를 전달한다. 
# 기울기: line_fitter.coef_
# 절편: line_fitter.intercept_
model = lr.fit(X_train, y_train)
```

```python
# 학습된 계수를 출력합니다.
print(lr.coef_)
```

```python
import statsmodels.api as sm
```

```python
# statsmodel 라이브러리로 회귀 분석을 수행합니다.
# 회귀분석모형 수식을 간단하게 만들기 위해 상수항을 독립변수 데이터에 추가하는 것을 상수항 결합 작업이라고 한다.
# add_constant() : 상수항 결합을 해주는 함수
X_train = sm.add_constant(X_train)
# OLS은 가장 기본적인 결정론적 회귀 방법으로 잔차제곱합(RSS)를 최소화하는 가중치 벡터를 행렬 미분으로 구하는 방법이다.
# OLS는 전처리한 자료에 회귀분석을 적용해 변수별 회귀계수를 구할 수 있다.
model = sm.OLS(y_train, X_train).fit()
```

```python
# summary() 분석에 대한 요약을 한눈에 볼 수 있다.
model.summary()
```

```python
# 한글 출력을 위한 사전 설정 단계입니다.
mpl.rc('font', family='NanumGothicOTF')
plt.rcParams['figure.figsize'] = [20,16]

# 회귀 계수를 리스트로 반환합니다.
coefs = model.params.tolist()
coefs_series = pd.Series(coefs)

# 변수명을 리스트로 반환합니다.
x_labels = model.params.index.tolist()
```

```python
# 회귀 계수를 출력합니다.
ax = coefs_series.plot(kind='bar')
ax.set_title('feautre_coef_graph')
ax.set_xlabel('x_features')
ax.set_ylabel('coef')
ax.set_xticklabels(x_labels)
```

```python
# 학습 데이터와 테스트 데이터로 분리합니다.
x = picher_df[picher_df.columns.difference(['선수명','y'])]
y = picher_df['y']
# train_test_split 함수는 전체 데이터셋 배열을 받아서 랜덤하게 test/train 데이터 셋으로 분리해주는 함수이다.
# Dataset을 나눌 때 test_size 옵션으로 Train, Test의 비율을 설정할 수 있고, random_state로 seed 값을 지정할 수 있다.
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
# 회귀 분석 모델을 학습합니다.
# linear_model.LinearRegression() : LinearRegression 모델을 생성
lr = linear_model.LinearRegression()

# fit() 메서드는 선형 회귀 모델에 필요한 두 가지 변수를 전달한다. 
# 기울기: line_fitter.coef_
# 절편: line_fitter.intercept_
model = lr.fit(X_train, y_train)
```

```python
# 회귀 분석 모델을 평가합니다.
print(model.score(X_train, y_train)) # train R2 score을 출력합니다.
print(model.score(X_test, y_test)) # test R2 score을 출력합니다.
```

```python
# 회귀 분석 모델을 평가합니다.
# predict() 함수는 주어진 새로운 변수(X)에 대한 예측을 수행한 결과를 예측값(fit – 점추정값에 해당)으로 출력한다. 
y_predictions = lr.predict(X_train)
print(sqrt(mean_squared_error(y_train, y_predictions))) # train RMSE score를 출력합니다.
y_predictions = lr.predict(X_test)
print(sqrt(mean_squared_error(y_test, y_predictions))) # test RMSE score를 출력합니다. 
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
# VIF(분선팽창요인)이 10~15 정도를 넘으면 다중 공선성 문제가 발생했다 판단한다.
# 피처마다의 VIF 계수를 출력합니다.
vif = pd.DataFrame()
vif["VIF Factor"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif["features"] = X.columns
# 소수점을 n번째 까지만 표현하고 반올림을 하고싶을때, round 함수를 사용하면된다.
vif.round(1)
```

```python
# 원래의 데이터 프레임을 다시 불러옵니다.
picher = pd.read_csv(picher_file_path)
picher = picher[['선수명', '연봉(2017)']]
```

```python
# 원래의 데이터 프레임에 2018년 연봉 정보를 합칩니다.
result_df = picher_df.sort_values(by=['y'], ascending=False)
# axis=0 은 '열' axis=1은 '행'
result_df.drop(['연봉(2017)'], axis=1, inplace=True, errors='ignore')
# pandas.merge() : 하나 이상의 키를 기준으로 DataFrame의 로우를 합침
# merge(데이터프레임, 데이터프레임, on = '기준')
result_df = result_df.merge(picher, on=['선수명'], how='left')
result_df = result_df[['선수명', 'y', '예측연봉(2018)', '연봉(2017)']]
result_df.columns = ['선수명', '실제연봉(2018)', '예측연봉(2018)', '작년연봉(2017)']
```

```python
# 재계약하여 연봉이 변화한 선수만을 대상으로 관찰합니다.
result_df = result_df[result_df['작년연봉(2017)'] != result_df['실제연봉(2018)']]
# 행 인덱스 초기화 : reset_index() 함수
result_df = result_df.reset_index()
# iloc	행 번호를 기준으로 행 데이터 읽기
result_df = result_df.iloc[:10, :]
```
